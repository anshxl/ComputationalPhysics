# Analysis of Coupled Oscillations in a Three-Pendulum System

### Describing the System

We have a system of three identical pendulums, each with mass $m$ and length $l$, coupled by springs with a spring constant $k$. This setup allows each pendulum to influence its neighbors. 

Each pendulum has a small angular displacement $\theta_i$ from the vertical, where $i = 1, 2, 3$. The coupling springs introduce forces that depend on the relative displacements between adjacent pendulums, leading to coupled oscillations.

#### Parameters:
- $m$: mass of each pendulum bob
- $l$: length of each pendulum
- $g$: acceleration due to gravity
- $k$: spring constant of the coupling springs

### Explaining the Approach

To analyze this system:
1. We’ll derive the equations of motion using **Lagrangian mechanics** by formulating the kinetic and potential energies.
2. Using the Lagrangian, we’ll obtain the **Euler-Lagrange equations** for each pendulum, resulting in coupled differential equations.
3. These equations will be arranged in **matrix form**, allowing us to find the normal modes of the system.


### Setting Up the Equations

#### 1. Kinetic Energy $ (T) $

For small displacements, the kinetic energy for each pendulum bob is:
$$
T = \frac{1}{2} m l^2 \left( \dot{\theta}_1^2 + \dot{\theta}_2^2 + \dot{\theta}_3^2 \right)
$$

#### 2. Potential Energy $ (V) $

The potential energy includes:
- **Gravitational potential energy**: $ \frac{1}{2} m g l \left( \theta_1^2 + \theta_2^2 + \theta_3^2 \right) $
- **Spring potential energy** due to coupling: $ \frac{1}{2} k \left( (\theta_1 - \theta_2)^2 + (\theta_2 - \theta_3)^2 \right) $

Thus, the total potential energy is:
$$
V = \frac{1}{2} m g l \left( \theta_1^2 + \theta_2^2 + \theta_3^2 \right) + \frac{1}{2} k \left( (\theta_1 - \theta_2)^2 + (\theta_2 - \theta_3)^2 \right)
$$

#### 3. Lagrangian $\mathcal{L}$

The Lagrangian $ \mathcal{L} = T - V $ is:
$$
\mathcal{L} = \frac{1}{2} m l^2 \left( \dot{\theta}_1^2 + \dot{\theta}_2^2 + \dot{\theta}_3^2 \right) - \frac{1}{2} m g l \left( \theta_1^2 + \theta_2^2 + \theta_3^2 \right) - \frac{1}{2} k \left( (\theta_1 - \theta_2)^2 + (\theta_2 - \theta_3)^2 \right)
$$

### Deriving Equations of Motion

Using the Euler-Langrange formulation, we get the equations of motion for the coupled pendulums:

1. For $ \theta_1 $:
   $$
   m l^2 \ddot{\theta}_1 + (m g l + k) \theta_1 - k \theta_2 = 0
   $$

2. For $ \theta_2 $:
   $$
   m l^2 \ddot{\theta}_2 + (m g l + 2k) \theta_2 - k (\theta_1 + \theta_3) = 0
   $$

3. For $ \theta_3 $:
   $$
   m l^2 \ddot{\theta}_3 + (m g l + k) \theta_3 - k \theta_2 = 0
   $$

### Matrix Representation

In matrix form, the equations of motion can be written as:

$$
m l^2 \begin{pmatrix} \ddot{\theta}_1 \\ \ddot{\theta}_2 \\ \ddot{\theta}_3 \end{pmatrix} + \begin{pmatrix} m g l + k & -k & 0 \\ -k & m g l + 2k & -k \\ 0 & -k & m g l + k \end{pmatrix} \begin{pmatrix} \theta_1 \\ \theta_2 \\ \theta_3 \end{pmatrix} = 0
$$

Or equivalently:

$$
\begin{pmatrix} \ddot{\theta}_1 \\ \ddot{\theta}_2 \\ \ddot{\theta}_3 \end{pmatrix} = -\frac{1}{m l^2} \begin{pmatrix} m g l + k & -k & 0 \\ -k & m g l + 2k & -k \\ 0 & -k & m g l + k \end{pmatrix} \begin{pmatrix} \theta_1 \\ \theta_2 \\ \theta_3 \end{pmatrix}
$$

This matrix formulation is ready for solving the eigenvalue problem to find the normal modes and frequencies.

### Finding the Normal Mode Frequencies

We will calculate the eigenvalues of the coefficient matrix from the matrix form of the equations of motion. The eigenvalues represent the squared angular frequencies $ \omega^2 $ of the system's normal modes.

To find the normal mode frequencies, we will:

1. Calculate the eigenvalues of the coefficient matrix.
2. Take the square root of each eigenvalue to find the corresponding angular frequency $\omega$.

The eigenvalues will give us the squared frequencies $\omega^2 $, and each eigenvalue corresponds to one of the normal modes of the system.

```python
import sympy as sp

# Define symbolic variables
m, l, g, k = sp.symbols('m l g k')

# Define the symbolic matrix
matrix = sp.Matrix([
    [m * g * l + k, -k, 0],
    [-k, m * g * l + 2 * k, -k],
    [0, -k, m * g * l + k]
]) / (m * l**2)

# Calculate the eigenvalues symbolically
eigenvalues = matrix.eigenvals()

# Display the eigenvalues
print("Eigenvalues (ω^2) for the normal modes:")
for eigenvalue, multiplicity in eigenvalues.items():
    sp.pretty_print(eigenvalue)
    print()
```

### Normal Mode Analysis

From our matrix formulation, we obtained the eigenvalues, which correspond to the squared angular frequencies $ \omega^2 $ of the normal modes of our system. These eigenvalues represent the characteristic frequencies at which the coupled pendulums oscillate in each mode. Each eigenvalue $ \lambda_i = \omega_i^2 $ describes a unique mode of vibration.

The eigenvalues (normal mode frequencies) we found are:

1. **Mode 1:**  
   $$
   \omega_1^2 = \frac{g}{l}
   $$
   This mode represents a frequency that depends solely on gravitational and pendulum length parameters, indicating a motion similar to that of a simple pendulum. In this mode, each pendulum may oscillate independently, with minimal influence from the coupling springs.

2. **Mode 2:**  
   $$
   \omega_2^2 = \frac{m g l + k}{l^2 m}
   $$
   This mode includes the effects of both gravity and the coupling constant $ k $. It represents a motion in which the coupling springs play a role in determining the oscillation frequency, suggesting that adjacent pendulums are now more interconnected in their motion.

3. **Mode 3:**  
   $$
   \omega_3^2 = \frac{m g l + 3 k}{l^2 m}
   $$
   This mode represents the highest frequency of oscillation among the three modes, with an increased influence from the coupling constant $ k $. The system’s response at this frequency indicates a strong interdependence between the pendulums due to the spring couplings.

### Interpretation of the Normal Modes

Each normal mode corresponds to a specific pattern of oscillation for the system:
- **Mode 1** suggests independent oscillations, similar to a simple pendulum.
- **Mode 2** indicates some level of interdependence between pendulums, as spring coupling becomes relevant.
- **Mode 3** represents the highest level of coupling, with all pendulums strongly influencing one another’s motion.

By calculating these frequencies, we gain insight into the natural oscillatory behavior of the system, allowing us to understand how the pendulums move collectively in each mode. This sets the foundation for visualizing or simulating these motions in the next section.

### Plotting the Normal Modes

To visualize the normal modes of our coupled pendulum system, we’ll plot the displacement of each pendulum as a function of time for each mode. Each mode corresponds to a unique frequency and amplitude pattern, and in each, the pendulums oscillate with characteristic phase relationships.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
m, l, g, k = 1.0, 1.0, 9.81, 0.5  # mass, length, gravity, spring constant
t = np.linspace(0, 20, 500)       # time array

# Eigenvalues (squared frequencies) and eigenvectors (mode shapes)
eigenvals = np.array([g / l, (m * g * l + k) / (m * l**2), (m * g * l + 3 * k) / (m * l**2)])
eigenvecs = np.array([
    [1, 0, -1],  # Mode 1
    [1, -2, 1],  # Mode 2
    [1, 1, 1]    # Mode 3
])

# Plot each mode
for mode in range(3):
    # Get mode frequency and eigenvector
    freq = np.sqrt(eigenvals[mode])
    amplitude = eigenvecs[mode]

    # Calculate displacements for each pendulum
    theta1 = amplitude[0] * np.cos(freq * t)
    theta2 = amplitude[1] * np.cos(freq * t)
    theta3 = amplitude[2] * np.cos(freq * t)

    # Plot results
    plt.figure(figsize=(10, 4))
    plt.plot(t, theta1, label=r'$\theta_1$', color='steelblue', alpha=0.8)
    plt.plot(t, theta2, label=r'$\theta_2$', color='orange', alpha=0.6)
    plt.plot(t, theta3, label=r'$\theta_3$', color='seagreen', alpha=0.5)
    plt.title(f'Normal Mode {mode + 1} (Frequency = {freq:.2f} rad/s)')
    plt.xlabel('Time')
    plt.ylabel('Displacement')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()
```

### Interpretation of Normal Modes

1. **Normal Mode 1 (Frequency = 3.13 rad/s):**
   - In this mode, \( \theta_1 \) and \( \theta_3 \) oscillate with equal amplitude but in opposite directions, while \( \theta_2 \) remains at rest (zero displacement).
   - This pattern suggests that the outer pendulums oscillate symmetrically about the central pendulum, which acts as a fixed pivot point.

2. **Normal Mode 2 (Frequency = 3.21 rad/s):**
   - Here, all three pendulums oscillate, but \( \theta_2 \) has a larger amplitude compared to \( \theta_1 \) and \( \theta_3 \), which oscillate in opposite directions.
   - This mode indicates an "in-phase" motion between the outer pendulums with a stronger response from the central pendulum, highlighting a collective oscillation where coupling effects play a significant role.

3. **Normal Mode 3 (Frequency = 3.36 rad/s):**
   - In this highest frequency mode, all three pendulums oscillate in sync with nearly identical amplitudes and phases.
   - This mode represents a uniform motion where the entire system oscillates together, maximizing the coupling effect.

Each mode reveals a different oscillatory behavior of the coupled pendulums, from independent to collective motion, demonstrating the impact of the coupling springs on the system dynamics.

### Numerical Solution 

We now define a numerical approach that will let us visualize the motion of the system for any chocie of initial conditions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters
m = 1.0      # Mass of each pendulum bob
l = 1.0      # Length of each pendulum
g = 9.81     # Gravitational acceleration
k = 0.5      # Spring constant for coupling

# Define the system of differential equations
def diffyqs(t, y, m, l, g, k):
    theta1, theta2, theta3, theta1_dot, theta2_dot, theta3_dot = y

    # Second derivatives based on equations of motion
    theta1_ddot = -(g / l + k / (m * l**2)) * theta1 + (k / (m * l**2)) * theta2
    theta2_ddot = (k / (m * l**2)) * theta1 - (g / l + 2 * k / (m * l**2)) * theta2 + (k / (m * l**2)) * theta3
    theta3_ddot = (k / (m * l**2)) * theta2 - (g / l + k / (m * l**2)) * theta3

    return [theta1_dot, theta2_dot, theta3_dot, theta1_ddot, theta2_ddot, theta3_ddot]

# Random initial conditions for the angles and velocities
np.random.seed(0)  # For reproducibility
initial_conditions = np.random.rand(6) * 0.2 - 0.1  # Small random angles and velocities

# Time span for the integration
t_span = (0, 20)
t_eval = np.linspace(0, 20, 500)

# Solve the system using solve_ivp
solution = solve_ivp(diffyqs, t_span, initial_conditions, args=(m, l, g, k), t_eval=t_eval, method='RK45')

# Extract solutions for each angle
theta1, theta2, theta3 = solution.y[0], solution.y[1], solution.y[2]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(solution.t, theta1, label=r'$\theta_1$')
plt.plot(solution.t, theta2, label=r'$\theta_2$')
plt.plot(solution.t, theta3, label=r'$\theta_3$')
plt.title("Time Evolution of the System for Given Initial Conditions")
plt.xlabel("Time")
plt.ylabel("Displacement")
plt.legend()
plt.grid()
plt.show()
```

### Exploring Synchronization

This model lacks any dissipative forces that would drive the system to select any three of the modes that would prove to be stable. Here, I try to add a damping force, which should allow the system to gradually lose energy, favoring the lowest-frequency mode over time, which can lead to synchronization.

```python
def diffyqs_damped(t, y, m, l, g, k, b):
    theta1, theta2, theta3, theta1_dot, theta2_dot, theta3_dot = y

    # Second derivatives with damping
    theta1_ddot = -(g / l + k / (m * l**2)) * theta1 + (k / (m * l**2)) * theta2 - (b / (m * l**2)) * theta1_dot
    theta2_ddot = (k / (m * l**2)) * theta1 - (g / l + 2 * k / (m * l**2)) * theta2 + (k / (m * l**2)) * theta3 - (b / (m * l**2)) * theta2_dot
    theta3_ddot = (k / (m * l**2)) * theta2 - (g / l + k / (m * l**2)) * theta3 - (b / (m * l**2)) * theta3_dot

    return [theta1_dot, theta2_dot, theta3_dot, theta1_ddot, theta2_ddot, theta3_ddot]

# Parameters
m = 1.0      # Mass of each pendulum bob
l = 1.0      # Length of each pendulum
g = 9.81     # Gravitational acceleration
k = 0.5      # Spring constant for coupling
b = 0.05      # Damping coefficient

# Random initial conditions for the angles and velocities
np.random.seed(0)  # For reproducibility
initial_conditions = np.random.rand(6) * 0.2 - 0.1  # Small random angles and velocities

# Time span for the integration
t_span = (0, 100)
t_eval = np.linspace(0, 100, 2500)

# Solve the system using solve_ivp
solution = solve_ivp(diffyqs_damped, t_span, initial_conditions, args=(m, l, g, k, b), t_eval=t_eval, method='RK45')

# Extract solutions for each angle
theta1, theta2, theta3 = solution.y[0], solution.y[1], solution.y[2]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(solution.t, theta1, label=r'$\theta_1$')
plt.plot(solution.t, theta2, label=r'$\theta_2$')
plt.plot(solution.t, theta3, label=r'$\theta_3$')
plt.title("Time Evolution of the System for Given Initial Conditions")
plt.xlabel("Time")
plt.ylabel("Displacement")
plt.legend()
plt.grid()
plt.show()
```

1. **Decay of Higher-Frequency Oscillations**: Initially, you see a mix of oscillations with varying frequencies and amplitudes, which is typical for random initial conditions exciting multiple modes. However, as time progresses, the amplitude of these oscillations steadily decreases, showing the effect of the damping term.

2. **Tendency Toward Synchronization**: Around the middle of the plot, the oscillations become more uniform in both frequency and amplitude. This suggests that the system is gradually favoring the lower-frequency modes, which is what we expect when damping is applied. 

3. **Emergence of the First Mode**: Towards the end, the system oscillates at a much lower amplitude and almost in sync, indicating it’s mostly oscillating in the first normal mode. This mode, being the lowest frequency, is the last to decay and thus becomes more pronounced over time as the higher-frequency modes dissipate.

In summary, this plot demonstrates how adding a small amount of damping helps in mode selection, with the system eventually favoring the lowest-frequency mode and exhibiting a synchronized oscillatory behavior. It’s a nice illustration of how dissipative effects promote coherence in coupled oscillatory systems.
